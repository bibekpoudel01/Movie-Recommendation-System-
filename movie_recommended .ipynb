```python
import ast
```

```python
def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies['cast'][0]
```

```python
def convert(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    return l
  
```

```python
movies['cast']=movies['cast'].apply(convert)
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x] )
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])


```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head(10)
```

```python
new_df['tags']=new_df['tags'].apply(lambda x: " ".join(x))
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x: x.lower() if isinstance(x, str) else "")
```

```python
new_df['tags'][0]
```

```python
new_df['tags'][1]
```

```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = model.encode(new_df['tags'].tolist(), batch_size=32, show_progress_bar=True)
```

```python
embeddings
```

```python
print(embeddings[0])
print(embeddings.shape)  # (4804, 384)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(embeddings)
```

```python
def recommend(movie_name, top_n=5):
    if movie_name not in new_df['title'].values:
        print("Movie not found!")
        return
    idx = new_df[new_df['title'] == movie_name].index[0]
    distances = list(enumerate(similarity[idx]))
    movies_list = sorted(distances, key=lambda x: x[1], reverse=True)[1:top_n+1]
    print(f"Movies similar to '{movie_name}':")
    for i in movies_list:
        print(new_df.iloc[i[0]].title)

  
```
